```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp = pd.read_csv('selection_params.txt')
print(temp)
```

```python
col = ['Unnamed: 0', 'Population', ' Number Synonymous',' Number Nonsynonymous', 
' Bin size', ' Slope', ' Intercept', ' pvalue','Chromosome']

pop = ['African','Latino','Ashkenazi Jewish','East Asian','European (Finnish)', 
'European (non-Finnish)','South Asian']
```

```python
for k in [' Slope', ' Intercept', ' pvalue']:
    plt.figure(figsize=(15,8))
    for p in pop:
        bypop = temp[(temp['Population']==p)] #& (temp[k]!=' nan')
        #print(bypop['Chromosome'])
        sns.scatterplot(bypop['Chromosome'], bypop[k].astype(float), label=p)
    plt.ylabel(k)
    plt.show()
```

```python
temp2 = pd.read_csv('fit_params.txt')
temp2.columns
#pd.unique(temp2[' Population'])
```

```python
for k in [' KS pvalue', ' KS Neutral pvalue']:
    for syn in pd.unique(temp2['Category']):
        plt.figure(figsize=(15,8))
        for p in pd.unique(temp2[' Population']):
            bypop = temp2[(temp2[' Population']==p)&(temp2['Category']==syn)] #& (temp[k]!=' nan')
            #print(bypop)
            #print(bypop['Chromosome'])
            sns.scatterplot(bypop['Chromosome'], bypop[k].astype(float), label=p)
        plt.title(syn+' mutations')
        plt.ylabel(k)
        plt.show()
```
